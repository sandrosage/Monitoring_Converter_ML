# DATA PREPROCESSING
## Recording Setup
    For the static data generation we tested the below dcdc-Converters. They were all tested with different loads (0A, 1A, 2A, 3A, 4A, 5A, 6A).
    115200 kbps UART, delay(2) excluded
    Every recording had a duration of 30 minutes.
    We also measured the room temperature to have a reference value.

    We used three different dcdc-Converter:
        - DCHY2400-08 (Normal)
        - DCHY2400-15 (manipulated with Drossel)
        - DCHY2400-12 (manipulated with Mosfets)
    
    You can find the recordings in the StaticData_Recordings directory. When the Preprocessing.clean_textfile() is called it stores the processed data back into the StaticData_Recordings directory.

    After this step, it starts with the real Processing for getting the right csv-files which are stored in the Preprocessed_StaticData_Recordings directory. 

    After executing the cell below you can start with using the .csv-Files.

```python
import pandas as pd
import os
from Preprocessing.data_preprocessing import Preprocessing    #import from self-written module

names=["vIn", "currentIn", "vOut", "onBoardTemp", "roomTemp"]   #these are the column names for the csv/dataframe
subfolder = ["drossel/", "mosfets/", "normal/"]  #subfolders within the source folder: necessary for iteratively get the data textfiles
source_folder = "Data/Static_Data/"   #source folder where all the data textfiles and recordings are

destination_folder = "Data/Preprocessed_StaticData_Recordings"   #folder where you want to store the processed data files

for folder in subfolder:
    Preprocessing.clean_textfile(source_folder=source_folder, folder=folder)    #function for cleaning the textfiles: f.ex. remove [ ; ( ; ]
    for idx, filename in enumerate(os.listdir(source_folder+folder)):
        print(source_folder + folder + filename)
        print(Preprocessing.convert_filename(source_folder + folder + filename))

        #final processing function which creates csv, divides the data by factor 2 and deletes wrong rows with missing data
        Preprocessing.prepCon_textfile(source_folder + folder + filename, column_names = names, folder = folder, destination_folder=destination_folder)
```
