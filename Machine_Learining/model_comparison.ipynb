```python
import os
import tensorflow as tf
import keras
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from sklearn.preprocessing import MinMaxScaler
from keras.losses import CategoricalCrossentropy
from keras.optimizers import Adam, RMSprop
```

```python
df = pd.read_csv("../static_merged_data.csv")
x = df.drop(labels="status", axis=1)
y = df[["status"]]
min_max_scale = MinMaxScaler()
x_scaled = min_max_scale.fit_transform(x)

def prepare_train_test_set(X, Y, num_classes):
    X_train, X_test, y_train, y_test = train_test_split(X, Y, shuffle=True, test_size=.2)
    train_labels = y_train.status.astype('category').cat.codes.to_numpy()
    test_labels = y_test.status.astype('category').cat.codes.to_numpy()
    train_labels = to_categorical(train_labels, num_classes)
    test_labels = to_categorical(test_labels, num_classes)
    return (X_train, X_test), (train_labels, test_labels)

(X_train, X_test), (train_labels, test_labels) = prepare_train_test_set(x_scaled,y,3)
```

```python
def transform_one_hot(df):
    df['status'] = df['status'].apply(lambda x: 0 if x == 'drossel' else (1 if x =="mosfets" else 2))
    return df
    
def transform(df, number):
    vector_array = np.ndarray(8,)
    for i in range(len(df)-1):
        newrow = np.append(df[i], df[i+1])
        vector_array = np.vstack([vector_array, newrow])
    labels = np.arange(vector_array.shape[0], dtype=int)
    labels = np.full_like(labels, number)
    return vector_array, labels

def vectorize(df):
    df = transform_one_hot(df)
    df_drossel = df[df["status"] == 0]
    df_drossel = df_drossel.drop("status", axis=1).to_numpy()
    df_mosfets = df[df["status"] == 1]
    df_mosfets = df_mosfets.drop("status", axis=1).to_numpy()
    df_normal = df[df["status"] == 2]
    df_normal = df_normal.drop("status", axis=1).to_numpy()
    drossel_features,drossel_labels = transform(df_drossel, 0)
    mosfets_features,mosfets_labels = transform(df_mosfets, 1)
    normal_features,normal_labels = transform(df_normal, 2)
    return np.concatenate((drossel_features, mosfets_features, normal_features)),np.hstack((drossel_labels, mosfets_labels, normal_labels))
```

```python
df = pd.read_csv("../static_merged_data.csv")
features, labels = vectorize(df)
X_train, X_test, y_train, y_test = train_test_split(features, labels, shuffle=True, test_size=.2)
train_labels = to_categorical(y_train, 3)
test_labels = to_categorical(y_test, 3)
```

```python
model_1 = Sequential([
    Input(shape=(8,), name="input-layer"),
    Dense(64, activation="relu"),
    Dense(64, activation="relu"),
    Dense(128, activation="relu"),
    Dense(256, activation="relu"),
    Dense(128, activation="relu"),
    Dense(16, activation="relu"),
    Dense(3, activation="softmax", name="output_layer")
])

model_1.compile(optimizer="adam", loss=CategoricalCrossentropy(), metrics=["accuracy"])

model_1.fit(X_train, train_labels,epochs=100, validation_split=0.1)
```
